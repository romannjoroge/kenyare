```python
import pandas as pd
import numpy as np

columns = [("Policy Holder ID", ""), ("Member ID", ""), ("Start Date of Cover", ""), ("End Date of Cover", ""), ("Date of Claim/Treatement Date", ""), ("Date of Payment", ""), ("Amount Paid", "Outpatient per family"), ("Amount Claimed", "Inpatient per family"), ("Amount Claimed", "Dental per individual"), ("Amount Claimed", "optic per individual"), ("Amount Claimed", "spectacle frame per individual"), ("Amount Claimed", "Death and total payment"), ("Amount Paid", "Outpatient per family"), ("Amount Paid", "Outpatient per family"), ("Amount Paid", "Inpatient per family"), ("Amount Paid", "Limit Dental per individual"), ("Amount Paid", "Limit optic per individual"), ("Amount Paid", "spectacle frame per individual"), ("Amount Paid", "Death and total"), ("Benefit Limit", "Total claims paid"), ("Benefit Limit", "Limit Outpatient"), ("Benefit Limit", "Limit Inpatient"), ("Benefit Limit", "Limit Dental"), ("Benefit Limit", "Limit optic"), ("Benefit Limit", "Limit spectacle"), ("Benefit Limit", "Death and total"), ("Provider Name/Code", "")]

df = pd.read_excel("data/Medical quota share treaty  3rd Qtr 2020.xlsx", header=[0,1], sheet_name="Claims Bordereaux", na_values=['###'])
df.fillna(np.nan)
df.head()
```

```python
# Clean column names to make them more sensible
import re
cols = df.columns
new_cols = []
for i, col in enumerate(cols):
    temp = col[1]
    temp = re.sub(r"Unnamed:.*", "", temp)
    new_cols.append((col[0], temp))
    
df.columns = pd.MultiIndex.from_tuples(new_cols)
df.head()
```

```python
# Fill NAs in policy holder ID with name of policy holder
df[('Policy Holder ID', '')] = df[('Policy Holder ID', '')].ffill()

df.head()
```

```python
# Get data for a specific cedant
cedant_df = df[df[('Policy Holder ID', '')] == "BICOR ASSURANCES GENERALES"]
cedant_df
```

```python
import datetime
import numpy as np

# Date of loss is within policy dates
df[("Date of Claim/Treatment Date", "")]

def isDateOfLossValid(date: str | None, startDate: datetime.date, endDate: datetime.date) -> bool:
    if(not isinstance(date, str)):
        return False
    convertedDate = datetime.datetime.strptime(date, '%Y-%m-%d').date()
    return not(convertedDate < startDate or convertedDate > endDate)

isDateOfLossValidSeries = df[("Date of Claim/Treatment Date", "")].apply(lambda x: isDateOfLossValid(x, startDate=datetime.date(year=2020, month=1, day=1), endDate=datetime.date(year=2020, month=12, day=31)))
isDateOfLossValidSeries

df=df[isDateOfLossValidSeries]

print("Invalid Date Of Loss")
invalid_date_loss_dataa = df[~isDateOfLossValidSeries]
invalid_date_loss_dataa.head()

def isDateInQuater(date: str | None, quater: int) -> bool:
    if(not isinstance(date, str)):
        return False
    convertedDate = datetime.datetime.strptime(date, '%Y-%m-%d').date()
    quaterOfDate = (convertedDate.month - 1) // 3 + 1
    return quaterOfDate == quater

isDateQuaterValid = df[("Date of Payment/Approval Date", "")].apply(lambda x: isDateInQuater(x, 3))
isDateQuaterValid
df=df[isDateQuaterValid]
invalid_date_quater_data=df[~isDateQuaterValid]

print("Invalid Date Quater")
invalid_date_quater_data.head()
```
